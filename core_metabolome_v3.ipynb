```python
import pandas as pd
import numpy as np
from rdkit.Chem.Descriptors import ExactMolWt
from rdkit.Chem.inchi import MolFromInchi
```

```python
cm2 = pd.read_pickle('/Users/dis/PycharmProjects/core_metabolome/core_metabolome_v2.pickle')
```

```python
cm2['Molecule'] = cm2.inchi.apply(lambda x: MolFromInchi(x))
```

```python
cm2 = cm2.dropna(subset=['Molecule'])
```

```python
cm2['exactmass'] = cm2.Molecule.apply(lambda x: ExactMolWt(x))
```

```python
cm2
```

```python
cm2 = cm2.sort_values(by='exactmass')
```

```python
cm2.to_csv('temp.csv', sep='\t')
```
